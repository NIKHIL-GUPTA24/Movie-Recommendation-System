```python
# Rename duplicate columns in credits
credits = credits.rename(columns={
    'movie_id': 'movie_id_cred',
    'cast': 'cast_cred',
    'crew': 'crew_cred'
})

# Merge with suffixes
merged_df = movies.merge(credits, on='title', suffixes=('_mov', '_cred'))
merged_df.head(2)
```

```python
movies = merged_df[['movie_id_cred','title','overview','genres','keywords','cast_cred','crew_cred']]
```

```python
import ast

def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(text):
    l=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter +=1

    return l
```

```python
movies['cast_cred'] = movies['cast_cred'].apply(convert_cast)
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        
        if i['job']=='Director':
            l.append(i['name'])
            break

    return l
```

```python
movies['crew_cred'] = movies['crew_cred'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast_cred']=movies['cast_cred'].apply(remove_space)
movies['crew_cred']=movies['crew_cred'].apply(remove_space)
movies['genres']=movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast_cred'] + movies['crew_cred']
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
new_df = movies[['movie_id_cred','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
new_df.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)    
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title']== movie].index[0]
    distances = sorted(list(enumerate(similarity [index])),reverse=True ,key= lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
